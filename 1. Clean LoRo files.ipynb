## Sport

```python
sport_xls = glob.glob("sport_files_loro/*.xls")
glob.glob("sport_files_loro/*.xls")
```

```python
# Parse through each Excel file

i_file = 0

for file in sport_xls: 
    file_name = os.path.basename(file)
    
    # Get their sheet names with simple pd.read + None
    
    df = pd.read_excel(file, None)
    sheet_names = list(df.keys())
    
    # Extract year from file name
    
    file_name_start = re.search('\d{4}', file_name).group()
    file_name_end = int(file_name_start) + 1
    
    # Iterate through sheets
        
    i_sheet = 0
    
    # Get only first 9 sheets (10 is a recap)
    for sheet in sheet_names[:9]:

        df_sheet = pd.read_excel(file, sheet_name=sheet, skiprows=range(1, 4), header=None, usecols = [0,1])
        df_sheet.columns = ['institution', 'montant']
        df_sheet['domaine'] = str(sheet)
        df_sheet['start'] = file_name_start
        df_sheet['end'] = file_name_end

        # First sheet starts the dataframe 
            ## Skip some bad lines
        if i_sheet == 0:
            df = df_sheet[1:-1]
        else:
            df = pd.concat([df, df_sheet[1:-1]])

        i_sheet += 1        

    if i_file == 0:
        df_sport = df
    else:
        df_sport = pd.concat([df_sport, df])

    i_file += 1
            
print("Total length: ", len(df_sport))
```
